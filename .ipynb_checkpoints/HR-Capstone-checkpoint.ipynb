Although it's considered highly unlikely, there are still no unique data that can determine whether the duplicated data really is duplicated caused by an error or coincidentally being the same. Ideally we need some insights from the company

We decided on eliminating those duplicated data concidering the chances to have a duplicate data is suppose to be very slim. However this conclusion are only justified by the fact that we cannot gather any insights from anyone since this is a open source data

```python
#'salary' data preprocessing to map the salary
mask = {'low' : 1,
        'medium' : 2,
        'high' : 3}

df0['salary_numeric'] = df0['salary'].map(mask)
df0['salary_numeric']
```

```python
#Eliminating duplicates
df0 = df0.drop_duplicates()
df0.duplicated().sum()
```

Not only they contain similar values in some parameters (like 'satisfaction_level', 'performance', and 'monthly_work_hour'), there are also some indicators showing that the data is not a miss entry. Some of them are

    * 'salary_numeric' being slightly higher meaning that employee that dedicating more than 7 years proportionally have more person having higher salary compared to the employee that works below 7 years

With this we conclude that this is not considered to be a miss entry, nor an outlier since they also holds no extreme value in other parameters

### Exploratory Data Analysis [EDA]
* Objective : Exploring stories behind the data and extracting insights from that informations

```python
#Getting the descriptive stats
df.describe()
```

```python
#Mapping each parameter and identifying multicolinearity 
sns.pairplot(df)
```

```python
sns.heatmap(df.corr(numeric_only = True), annot = True, fmt = '.2f')
```

There are no variable that are highly correlated one to another meaning there are no multicolinearity detected. In other words all variables are independent to each other.

One last thing to clarify is whether this turnover happen in 1 particular department or not.

```python
pd.crosstab(df['department'], df['left'], normalize = 'index')
```

Mainly, each department hold a similar turnover rate except for management which is completely understandable since a good management are preferably people with good amount of experience in that area. Other than that this data shows that 'department' doesn't seems to have a good relationship with the person's status of left or not.

```python
#Grouping data by 'left'
df.groupby('left').describe()
```

Mainly, employee who left and employee who didn't left have similar descriptive statistics. There are 2 notable difference that is 'satisfaction_level' which is lower for employee who left, and promotion_last_5years which is also lower for employee who left.

To get more clarity, we compared 'satisfaction_level' and 'performance' to determine whether the best employee really satisfied by their job.


```python
#plotting 'satisfaction_level' vs 'performance'
plt.figure(figsize = (15, 8))
sns.scatterplot(data = df, x = 'satisfaction_level', y = 'performance', hue = 'left')
plt.title('Satisfaction_level vs Performance scores')
```

Surprisingly, the data forming a 2 group which is 

    - Group 1 ('satisfaction_level' below 0.110)
    
    - Group 2 ('satisfaction_level between 0.3-0.5 and 'performance between 0.4-0.6)

```python
df.info()
```

```python
#Splitting dataset into groups according to the trend detected
group1 = df.loc[(df['satisfaction_level'] <= 0.110) & 
                (df['left'] == 1)]

group2 = df.loc[
    ((df['satisfaction_level'] < 0.5) & (df['satisfaction_level'] > 0.3)) &
    ((df['performance'] > 0.4) & (df['performance'] < 0.6)) &
    (df['left'] == 1)
]

group1.head()
```

**Analysis for group 1**

```python
group1.describe()
```

```python
df[~df.index.isin(group1.index)].describe()
```

Compared to the other employee, those who categorized as group1 are far more superior in terms of 'performance', 'number_project', 'monthly_work_hour', and 'promotion_last_5years'

In contrast, as we can see employee that belongs to group1 has very low 'satisfaction_level', and 'work_accident'. One thing to note is that 'salary_numeric' of group1 are lesser compared to those outside of group1.

> There are indication of miss management where people with great performance are not rewarded enough. They work 275 hours in a month (about 11 hours a day [25 days in 1 month]) and work 6 projects. Almost none of them gets promotion, and there are less employee that receive a good salary compared to the group outside of this group. 


```python
#Check whether this mistakes happen in a specific department or all employee in general
df.loc[(df['satisfaction_level']<=0.110) & (df['left'] == 1)].groupby('department').size().sort_values(ascending = False)
```

```python
#Checking the proportions (how many percent of the employee who left belongs to this group)
(df.loc[(df['satisfaction_level']<=0.110) & (df['left'] == 1)].groupby('department').size() / df[df['left'] == 1].groupby('department').size()).sort_values(ascending = False)
```

This miss manaement are seems to be occuring not on specific department but happen on all department. This problem particularly happen more often in department with high skill requirement (IT, technical, RandD, accounting, and management). Since it is more difficult looking for these people with high skill requirement, it is adviced to reconsider rewards for those who works better to prevent potentially capable employee leaving the company.

```python
#Checking 'time_spend_company' since 25th, 50th, and 75th percentiles are all 4 years
group1.groupby('time_spend_company').size()
```

This group is also dominated by employee who quits after 4 years. This might be a clue to an accident causing massive turnover on great employee who works really hard.
